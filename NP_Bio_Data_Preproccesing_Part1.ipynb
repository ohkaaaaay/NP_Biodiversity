## Step 2: Join National Park Visitor Count to Parks Table

```python
# Check the datatypes of the columns in visits
visits.dtypes
```

```python
# Modify ParkName to reflect the same format in parks table
# Replace NP & PRES
visits['ParkName'] = visits.ParkName.str.replace('NP & PRES','National Park and Preserve')
# Check visits table
visits
```

```python
# Modify ParkName to reflect the same format in parks table
# Replace NP
visits['ParkName'] = visits.ParkName.str.replace('NP','National Park')
# Check visits table
visits
```

### Year Selection
**Note on Kaggle**: "Plant and animal species in parks (downloaded 6 January 2017)"

**Action**: Filter Total Record Visitors (TRV) from 2016.

```python
visit2016 = visits[visits["Year"]==2016]
# Check visits in 2016
visit2016
```

```python
# Modify visit2016 to select only the needed columns
visit2016 = visit2016[['ParkName', 'TRV']]
# Check modified visits in 2016
visit2016
```

```python
species = pd.merge(species, visit2016, how='left', left_on='Park Name', right_on='ParkName')
species
```

### Remove Extra Park Name Column (ParkName)

```python
# Drop ParkName
species.drop('ParkName', inplace=True, axis=1)
```

```python
# Check species
species
```

```python
# Check species shape
species.shape
```

## Step 5: Check Distinct Values
Need to check distinct values for the following columns:
- Category
- Order
- Family
- Record Status
- Occurrence
- Nativeness
- Abundance
- Seasonality
- Conservation Status

```python
def uniqcol(sp, col):
    return sp[col].nunique()

def cncol(sp, col):
    return sp[col].value_counts()
```

### Category

```python
print(cncol(sp,'Category'))
print('Length:', uniqcol(sp,'Category'))
```

### Order

```python
print(cncol(sp,'Order'))
```
